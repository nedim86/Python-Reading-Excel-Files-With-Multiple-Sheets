# Reading Excel Files With Multiple Sheets

```python
import pandas as pd
```

```python
# We have a excel file big_mart_sales_with_multiple_sheets.xlsx that contain different sheets for different years 1985, 1987, 1997
# Read the data with multiple sheets
data_with_multiple_sheets = pd.read_excel('big_mart_sales_with_multiple_sheets.xlsx')
```

```python
# We can see that when we read the data with multiple sheets only the first sheet gets read.
data_with_multiple_sheets.shape
```

```python
# To read the data from the specific sheet, we need to specify the sheet name.

sheet_1985 = pd.read_excel('big_mart_sales_with_multiple_sheets.xlsx',sheet_name='1985')
sheet_1987 = pd.read_excel('big_mart_sales_with_multiple_sheets.xlsx', sheet_name='1987')
sheet_1997 = pd.read_excel('big_mart_sales_with_multiple_sheets.xlsx', sheet_name='1997')
```

```python
sheet_1985.shape, sheet_1987.shape, sheet_1997.shape
```

```python
# The next step is to concatenate all sheets (create an array of dataframes)
all_dataframes = [sheet_1985, sheet_1987, sheet_1997]
all_dataframes
```

```python
type(all_dataframes)
```

```python
len(all_dataframes)
```

```python
final_data = pd.concat(all_dataframes)
```
